# 1. Show the evolution of total monthly and total annual exports from Brazil (all states and to everywhere) of...

```python
#Adding Month and Year Reference Columns
df['month'] = df['date'].dt.to_period('M')
df['year'] = df['date'].dt.to_period('Y')
df.head()
```

```python
def plot_exports_by_period(product, period, df, color):

    #Allowed inputs
    allowed_products = ['corn', 'soybean_meal', 'soybean_oil', 'soybeans', 'sugar', 'wheat']
    allowed_periods = ['month', 'year']
    
    #check if function inputs are correct
    if not product in allowed_products or not period in allowed_periods:
        print("Invalid product or period inputs!")
        return 
    
    #This determines how frequently labels are displayed in the graph.
    if period == 'month':
        SHOW_LABEL_EVERY = 5
    else:
        SHOW_LABEL_EVERY = 1

    
    #filtering main df to target product
    filtered_df = df[df['product']==product]
    filtered_df = filtered_df.reset_index()

    #groups by chosen period
    grouped_sum = pd.DataFrame(filtered_df.groupby(period)["tons"].sum())
    
    #Parsing index to datetime
    grouped_sum = grouped_sum.reset_index()
    grouped_sum[period] = grouped_sum[period].values.astype('datetime64[M]')

    #setting up the graph
    fig, ax = plt.subplots(1,1, figsize = (15,7))
    fig.suptitle(f"{product.capitalize()} Exports by {period.capitalize()} (Million tons)")
    
    ax.plot(grouped_sum[period], grouped_sum['tons'], c=color)

    #setting y scale to million tons
    ax.get_yaxis().set_major_formatter(
        FuncFormatter(lambda x, p: format(int(x)/1_000_000, ',')))

    #Adding labels to the graph
    counter = 0 #used only to determine how frequently the labels are displayed.
    for x,y in zip(grouped_sum[period],grouped_sum['tons']):
        if counter % SHOW_LABEL_EVERY != 0:
            counter+=1
            continue

        label = "{:.2f}".format(y/1_000_000)

        plt.annotate(label, # text
                     (x,y), # text coordinates
                     textcoords="offset points", # text position
                     xytext=(0,10), # offset distance from text to points (x,y)
                     ha='center') # horizontal alignment -> [left, right, center]
        counter+=1
```

## 1.1. Soybeans

```python
plot_exports_by_period(product='soybeans', 
                       period='month', 
                       df=df, 
                       color='b')

plot_exports_by_period(product='soybeans', 
                       period='year', 
                       df=df, 
                       color='b')
```

## 1.2. Soybean oil

```python
plot_exports_by_period(product='soybean_oil', 
                       period='month', 
                       df=df, 
                       color='g')

plot_exports_by_period(product='soybean_oil', 
                       period='year', 
                       df=df, 
                       color='g')
```

## 1.3. Soybean Meal

```python
plot_exports_by_period(product='soybean_meal', 
                       period='month', 
                       df=df, 
                       color='y')

plot_exports_by_period(product='soybean_oil', 
                       period='year', 
                       df=df, 
                       color='y')
```

# 2. What are the 3 most important products exported by Brazil in the last 5 years?

## 2.1. By Volume

```python
def filter_last_n_years(df, n_years):
    filtered_df = df[df['year']>=df['year'].max()- (n_years-1)] #N year diff-1
    filtered_df = filtered_df.reset_index()
    return filtered_df
```

```python
#Filtering the last 5 years
last_5_years_df = filter_last_n_years(df, 5)
last_5_years_df.head()
```

```python
#Analyzed years
last_5_years_df['year'].unique()
```

```python
#Grouping total tons by product and selecting top 3
grouped_sum = pd.DataFrame(last_5_years_df.groupby("product")["tons"].sum())
grouped_sum = grouped_sum.sort_values(by='tons', ascending=False)
grouped_sum = grouped_sum.reset_index()
grouped_sum.nlargest(3, 'tons')
```

```python
fig, ax = plt.subplots(1,1, figsize = (15,7))
ax = sns.barplot(x="product", y="tons", data=grouped_sum)
fig.suptitle("Total exports by Products 2015-2019 (Million Tons)")

ax.get_yaxis().set_major_formatter(
    FuncFormatter(lambda x, p: format(int(x)/1_000_000, ',')))

for index, row in grouped_sum.iterrows():
    ax.text(row.name,row['tons'], round(row['tons']/1_000_000,2), color='black', ha="center")
```

### <font color='blue'>A: Soybeans (327.62 MT), Corn (151.42 MT) and Sugar (120.04 MT)</font>

# 2.2. By USD

```python
#Grouping total tons by product and selecting top 3
grouped_sum = pd.DataFrame(last_5_years_df.groupby("product")["usd"].sum())
grouped_sum = grouped_sum.sort_values(by='usd', ascending=False)
grouped_sum = grouped_sum.reset_index()
grouped_sum.nlargest(3, 'usd')
```

```python
fig, ax = plt.subplots(1,1, figsize = (15,7))
ax = sns.barplot(x="product", y="usd", data=grouped_sum)
fig.suptitle("Total exports by Products 2015-2019 (USD)")

ax.get_yaxis().set_major_formatter(
    FuncFormatter(lambda x, p: format(int(x)/1_000_000, ',')))

for index, row in grouped_sum.iterrows():
    ax.text(row.name,row['usd'], "{:,}".format(round(row['usd']/1_000_000,2)), color='black', ha="center")
```

### <font color='blue'>A: Soybeans (123,746.66 M USD), Sugar (40,947.32 M USD) and Soybean Meal (28,413.61 M USD)</font>

# 3.1. What are the main routes through which Brazil have been exporting ‘corn’ in the last few years?

```python
#Filtering Corn data of the last 5 years
corn_df = df[df['product']=='corn'].reset_index()
corn_df = filter_last_n_years(corn_df, 5)
corn_df.head()
```

```python
#Grouping total tons by product and selecting top 3
grouped_sum = pd.DataFrame(corn_df.groupby("route")["tons"].sum())
grouped_sum = grouped_sum.sort_values(by='tons', ascending=False)
grouped_sum = grouped_sum.reset_index()
grouped_sum
```

```python
#Plotting Volume by route
fig, ax = plt.subplots(1,1, figsize = (15,7))
ax = sns.barplot(x="route", y="tons", data=grouped_sum)
fig.suptitle("Total exports by Products 2015-2019 (Million Tons)")

ax.get_yaxis().set_major_formatter(
    FuncFormatter(lambda x, p: format(int(x)/1_000_000, ',')))

for index, row in grouped_sum.iterrows():
    ax.text(row.name,row['tons'], round(row['tons']/1_000_000,2), color='black', ha="center")
```

### <font color='blue'>A: The main route is definitely by Sea with minor contributions of 'River' and 'Ground route'. </font>

# 3.2. Are there differences in the relative importance of routes depending on the product?

```python
#Grouping by Product and route
grouped_sum = last_5_years_df.groupby(['product', 'route'])['tons'].sum().unstack('route').fillna(0)
grouped_sum
```

```python
#Plotting exported volumes by product and route
fig, ax = plt.subplots(1,1, figsize = (15,7))
fig.suptitle("Export Volumes by Product by Route")

grouped_sum[unique_routes].plot(kind='bar', stacked=True, ax=ax)
```

### <font color='blue'>A: As we can see on the graph, the same pattern is found on every product. So no, there are no differences in the relative importance of routes depending on the product.</font>

# 5. For each of the products in the dataset, show the 5 most important states in terms of exports.

### <font color='green'>Previous analysis showed a considerable decrease on the total exports of some products in the past few years. So I'll be considering only the last 5 years.</font>

```python
#Grouping by Product and state
grouped_sum = last_5_years_df.groupby(['product', 'state'])['tons'].sum().reset_index()
grouped_sum
```

```python
top_5_states_by_product = {}

for product in unique_products:
    product_df = grouped_sum[grouped_sum['product']==product].reset_index()
    top_5_states_by_product[product] =  product_df.nlargest(5, 'tons').reset_index()
```

```python
top_5_states_by_product['soybeans']
```

```python
#Grouping by Product and route

for product in unique_products:

    grouped_sum = top_5_states_by_product[product]

    top_states = list(top_5_states_by_product[product]['state'])

    #Plotting exported volumes by product and route
    fig, ax = plt.subplots(1,1, figsize = (15,7))
    ax = sns.barplot(x="state", y="tons", data=grouped_sum)
    fig.suptitle(f"Top 5 States total exports of {product.capitalize()} 2015-2019 (Million Tons)")

    ax.get_yaxis().set_major_formatter(
        FuncFormatter(lambda x, p: format(int(x)/1_000_000, ',')))

    for index, row in grouped_sum.iterrows():
        ax.text(row.name,row['tons'], "{:,}".format(round(row['tons']/1_000_000,2)), color='black', ha="center")
```

## Question: What should be the total brazilian soybeans, soybean_meal, and corn export forecasts, in tons, for the next 11 years (2020-2030)? Weâ€™re mostly interested in the annual forecast

```python
#Now we get the covariates data
covariates_df = pd.read_excel('covariates.xlsx')
covariates_df.head()
```

#### Since our covariates dataset contains basically economic growth rates of many countries and commodities prices, we should select only the countries that are most important in terms of export volume regarding each product

As seen on previous analysis, we should consider a recent time frame due to decreases on exports depeding upon the product. I'll be using the last 5 years to determine the most important destinations.

```python
#Reading data
exports_df = pd.read_csv('data_comexstat.csv', encoding='latin-1')

#Parsing date column to datetime
exports_df['date'] = pd.to_datetime(exports_df['date'], format='%Y-%m-%d')

#Adding Year Reference Column
exports_df['year'] = exports_df['date'].dt.to_period('Y')
exports_df['year'] = exports_df['year'].astype(str).astype(int)

#Grouping by total exports by year
#exports_df = exports_df.groupby(['product', 'year'])['tons'].sum().reset_index()

display(exports_df.head())
```

#### The exports dataset starts at 1997 whilst the covariates dataset starts at 1979. We should delete the rows that do not have a matching period.

```python
#Dropping rows before 1997
covariates_df = covariates_df[covariates_df['year']>=1997]
covariates_df['year'] = covariates_df['year'].astype(str).astype(int) #parsing to join with covariates df

#Index by year
covariates_df = covariates_df.set_index('year')

display(covariates_df.head())

display(covariates_df.tail(11)) #showing next 11 years data as required
```

```python
def filter_last_n_years(df, n_years):
    filtered_df = df[df['year']>=df['year'].max()- (n_years-1)] #N year diff-1
    filtered_df = filtered_df.reset_index(drop=True)
    return filtered_df
```

```python
def filter_df_by_product(df, product):
    filtered_df = df[df['product'] == product].reset_index(drop=True)
    return filtered_df
```

```python
def last_n_years_top_n_destinations_df(exports_df, product, n_years, top_n):
    last_n_years_exports_df = filter_last_n_years(exports_df,n_years)
    last_n_years_product_df = filter_df_by_product(last_n_years_exports_df, product)
    grouped_sum = pd.DataFrame(last_n_years_product_df.groupby("country")["tons"].sum())
    top_n_destinations_df =  grouped_sum.nlargest(top_n, 'tons').reset_index()
    return top_n_destinations_df
```

# 1. Soybeans

```python
top_5_soybean_destinations_df = last_n_years_top_n_destinations_df(exports_df=exports_df, 
                                                                   product='soybeans', 
                                                                   n_years=5, 
                                                                   top_n=5)
top_5_soybean_destinations_df.head()
```

#### These are the top destinations of Brazillian Soybeans. Now we have to check if there's multicollinearity between the growth of each country or between commodities prices.

First, let's create a way to visualize the total annual exports of each product and concatenate the data to our covariate dataset.

```python
def get_annual_exports_by_product(exports_df, product):
    product_df = exports_df[exports_df['product']==product]
    grouped_sum = pd.DataFrame(product_df.groupby("year")["tons"].sum())
    grouped_sum = grouped_sum.reset_index()
    return grouped_sum
```

```python
def plot_annual_exports(product_df, color):
    fig, ax = plt.subplots(1,1, figsize = (15,7))
    #fig.suptitle(f"{product.capitalize()} Exports by {period.capitalize()} (Million tons)")
    
    
    ax.plot(product_df['year'], product_df['tons'], c=color)

    #setting y scale to million tons
    ax.get_yaxis().set_major_formatter(
        FuncFormatter(lambda x, p: format(int(x)/1_000_000, ',')))

    #Adding labels to the graph
    for x,y in zip(product_df['year'],product_df['tons']):
        label = "{:.2f}".format(y/1_000_000)

        plt.annotate(label, # text
                     (x,y), # text coordinates
                     textcoords="offset points", # text position
                     xytext=(0,5), # offset distance from text to points (x,y)
                     ha='center') # horizontal alignment -> [left, right, center]
```

```python
def concatenate_product_covariate_df(product_df, covariates_df):
    product_df = product_df.set_index('year')
    full_df = pd.concat([product_df, covariates_df], axis=1)
    return full_df
```

```python
#Soybeans annual exports
soybean_annual_exports = get_annual_exports_by_product(exports_df, 'soybeans')
display(soybean_annual_exports.tail())
plot_annual_exports(soybean_annual_exports, color='b')
```

```python
full_soybean_df = concatenate_product_covariate_df(soybean_annual_exports, covariates_df)

display(full_soybean_df.head())
display(full_soybean_df.tail(11)) #displaying years to be forecast
```

## Checking for multicollinearity

```python
# calculate the correlation matrix
corr = full_soybean_df.loc[:'2019'].corr() #filtering out the years that do not contain export data

# display the correlation matrix
display(corr)

# plot the correlation heatmap
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, cmap='RdBu')
```

### As we can see on the correlation matrix, many of the 'gdp measures' show high levels of correlation which is an indication of multicollinearity. We can double check it by the Variance inflation factor.

```python
# define two data frames one before the drop and one after the drop
full_soybean_df_before = full_soybean_df.loc[:'2019'] #filtering out the years that do not contain export data

# the VFI does expect a constant term in the data, so we need to add one using the add_constant method
X1 = sm.tools.add_constant(full_soybean_df_before)

# create the series for both
series_before = pd.Series([variance_inflation_factor(X1.values, i) for i in range(X1.shape[1])], index=X1.columns)

# display the series
display(series_before)
```

A general recommendation is that if any of our variables come back with a **value of 5 or higher, then they should be removed from the model.**

### As we can see, there are many variables with high values of Variance inflation factor. If we drop the highly correlated columns we should see a decrease on these numbers.

```python
'''
Soybeans Top Destinations

China
Spain
Thailand
Netherlands
Iran

'''

columns_to_keep = [
    'gdp_china',
    'gdp_spain',
    'gdp_thailand',
    'gpd_netherlands',
    'gdp_iran',
    
    'price_soybeans',
    'price_corn',
]

columns_to_drop = list(set(full_soybean_df.columns).symmetric_difference(set(columns_to_keep)))


full_soybean_df_after = full_soybean_df.loc[:'2019'].drop(columns_to_drop, axis = 1)

X2 = sm.tools.add_constant(full_soybean_df_after)

series_after = pd.Series([variance_inflation_factor(X2.values, i) for i in range(X2.shape[1])], index=X2.columns)

display(series_after)
```

### The numbers decreased a lot as expected, but are still too high (Ideally we should have numbers <= 5). After some trial and error, the optimal combination of variables for soybeans is:

```python
'''
Soybeans Top Destinations

China
Spain
Thailand
Netherlands
Iran

'''

columns_to_keep = [
    'gdp_china',
    #'gdp_spain',
    #'gdp_thailand',
    #'gpd_netherlands',
    #'gdp_iran',
    
    'price_soybeans',
    #'price_corn',
]

columns_to_drop = list(set(full_soybean_df.columns).symmetric_difference(set(columns_to_keep)))


full_soybean_df_after = full_soybean_df.loc[:'2019'].drop(columns_to_drop, axis = 1)

X2 = sm.tools.add_constant(full_soybean_df_after)

series_after = pd.Series([variance_inflation_factor(X2.values, i) for i in range(X2.shape[1])], index=X2.columns)

display(series_after)
```

##### We'll use as input of our models the variables: price_soybeans, gdp_china and gdp_spain

# Building the Model

```python
# define our input variable (X) & output variable
columns_to_drop.remove('tons') #we do not need to delete our target value now

full_soybean_df_after = full_soybean_df.loc[:'2019'].drop(columns_to_drop, axis = 1)

X = full_soybean_df_after.drop('tons', axis = 1)
Y = full_soybean_df_after[['tons']]

# Split X and y into X_
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=1)
```

```python
# define our input
X2 = sm.add_constant(X)

# create a OLS model
model = sm.OLS(Y, X2)

# fit the data
est = model.fit()
```

```python
# print out a summary
print(est.summary())
```

```python
# Get multiple predictions
y_predict = est.predict(sm.add_constant(X_test))

predicted_years = y_test.index.tolist()

# Show the first 5 predictions
for index, point in enumerate(y_predict):
    observed_value = round(y_test.iloc[index]['tons']/1000000, 2)
    predicted_value = round(point/1000000, 2)
    print(f" Year {predicted_years[index]} | Model prediction: {predicted_value} M tons | Observed value: {observed_value} | M tons | Dif: {round(observed_value-predicted_value,2)}")
#y_predict[:5]
```

### As we can see, the model does not show a lot of difference when predicting the data within the test sample. We can proceed with the prediction of the next 11 years now.

```python
soybeans_predict_period_df = full_soybean_df.loc['2020':].drop(columns_to_drop, axis = 1)
X_forecast = soybeans_predict_period_df.drop('tons', axis = 1)
X_forecast
```

```python
X_forecast = sm.add_constant(X_forecast)
final_forecast = pd.DataFrame(est.predict(X_forecast), columns=['tons'])
final_forecast = final_forecast.reset_index()
```

```python
plot_annual_exports(final_forecast, 'b')
```
